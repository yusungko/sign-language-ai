```python
import cv2
import mediapipe as mp
import numpy as np
import time
```

```python
def producedvideo(classname,number,path,dominator,numerator):
    handsFrameCount = 1
    percentage = int((numerator-dominator)*(100/numerator))
    deleteCount = 0
    skipCount = 0
    deleteHandsCounter = 1
    skipHandsCounter = 0
    
    if(dominator==0):
        deleteHandsCounter = 0
        skipHandsCounter = 1
    elif(dominator>1):
        deleteCount = dominator
        while(((numerator-skipCount)%dominator)!=0):
            skipCount+=1
        numerator=(numerator-skipCount)/dominator
        deleteHandsCounter = 0
    
    cap = cv2.VideoCapture(path)#'videos/02961.mp4'
    mpHands = mp.solutions.hands
    hands = mpHands.Hands(max_num_hands=2,model_complexity=1, min_detection_confidence=0.5, min_tracking_confidence=0.6)
    mpDraw = mp.solutions.drawing_utils
    handLandmarksStyle = mpDraw.DrawingSpec(color=(255, 0, 0), thickness=2)
    handConnectionStyle = mpDraw.DrawingSpec(color=(0, 255, 0), thickness=3)
    cTime = 0
    pTime = 0

    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))

    handsname = classname+'_'+"hands"+"_"+str(number)+"_"+str(percentage)+".mp4"
    reversedhandsname = classname+'_'+"hands"+"_"+str(number+20)+"_"+str(percentage)+".mp4"
    
    out1 = cv2.VideoWriter('videoss/'+handsname,cv2.VideoWriter_fourcc('m','p','4','v'), 10, (frame_width,frame_height))
    out4 = cv2.VideoWriter('videoss/'+reversedhandsname,cv2.VideoWriter_fourcc('m','p','4','v'), 10, (frame_width,frame_height))

    black_img = np.zeros((frame_height, frame_width, 3), dtype = np.uint8)
    frame1 = np.zeros((frame_height, frame_width, 3), dtype = np.uint8)
    frame4 = np.zeros((frame_height, frame_width, 3), dtype = np.uint8)

    while True:
        ret, img = cap.read()
        if ret:
            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            result = hands.process(imgRGB)

            imgHeight = img.shape[0]#*3
            imgWidth = img.shape[1]#*3
            #img = cv2.resize(img, None, fx=3, fy=3, interpolation = cv2.INTER_LINEAR)

            
            if result.multi_hand_landmarks:
                if(deleteHandsCounter!=deleteCount):
                    if((handsFrameCount%numerator)==0):
                        handsFrameCount+=1
                        out1.write(black_img)
                        out4.write(black_img)
                        deleteHandsCounter+=1
                    else:
                        for num, handLandmarks in enumerate(result.multi_hand_landmarks):
                            mpDraw.draw_landmarks(frame1, handLandmarks, mpHands.HAND_CONNECTIONS, handLandmarksStyle, handConnectionStyle)
                            for i, lm in enumerate(handLandmarks.landmark):
                                lm.x = 1-lm.x
                            mpDraw.draw_landmarks(frame4, handLandmarks, mpHands.HAND_CONNECTIONS, handLandmarksStyle, handConnectionStyle)
                        handsFrameCount+=1
                        out1.write(frame1)
                        frame1 = np.zeros((imgHeight, imgWidth, 3), dtype = np.uint8)
                        out4.write(frame4)
                        frame4 = np.zeros((imgHeight, imgWidth, 3), dtype = np.uint8)

                else:
                    for num, handLandmarks in enumerate(result.multi_hand_landmarks):
                        mpDraw.draw_landmarks(frame1, handLandmarks, mpHands.HAND_CONNECTIONS, handLandmarksStyle, handConnectionStyle)
                        for i, lm in enumerate(handLandmarks.landmark):
                            lm.x = 1-lm.x
                        mpDraw.draw_landmarks(frame4, handLandmarks, mpHands.HAND_CONNECTIONS, handLandmarksStyle, handConnectionStyle)
                    handsFrameCount+=1
                    out1.write(frame1)
                    frame1 = np.zeros((imgHeight, imgWidth, 3), dtype = np.uint8)
                    out4.write(frame4)
                    frame4 = np.zeros((imgHeight, imgWidth, 3), dtype = np.uint8)
                    skipHandsCounter+=1
                    if(skipHandsCounter==skipCount):
                        skipHandsCounter=0
                        deleteHandsCounter=0
                        
            else:
                out1.write(black_img)
                out4.write(black_img)
        else:
            
            print ("Finished Producing")
            break

    black_img = None
    del black_img
    frame1 = None
    del frame1
    frame4 = None
    del frame4
    cap.release()
    cv2.destroyAllWindows()
    out1.release()
    out4.release()
```
